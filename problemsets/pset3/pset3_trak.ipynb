B. Write a function that takes one press release string as an input and:

```python
def func(text):
    sent_obj = SentimentIntensityAnalyzer()
    sub = nlp(text)
    entities = set(ent.text for ent in sub.ents)
    
    cleaned_text = text
    
    for ent in entities:
        pattern = re.escape(ent)  # make it safe
        cleaned_text = re.sub(pattern, '', cleaned_text)

    sentiment = sent_obj.polarity_scores(cleaned_text)
    return sentiment
```

```python
custom_doj_stopwords = ["civil", "rights", "division", "department", "justice",
                        "office", "attorney", "district", "case", "investigation", "assistant",
                       "trial", "assistance", "assist"]
```

```python
## your code defining a text processing function
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
import re

default_stopwords = set(stopwords.words('english'))
all_stopwords = default_stopwords.union(custom_doj_stopwords)

stemmer = SnowballStemmer("english")

def preprocess_text(text):
    text = text.lower()

    tokens = re.findall(r'\b\w+\b', text)

    processed = [
        stemmer.stem(word) for word in tokens
        if word.isalpha() and word not in all_stopwords and len(word) >= 4
    ]

    return " ".join(processed)
```

```python
## your code executing the function
doj_subset_wscore["processed_text"] = doj_subset_wscore["contents"].apply(preprocess_text)
```

D. Print the top 10 words for press releases in each of the three `topics_clean`

For steps B - D, to receive full credit, write a function `get_topwords` that helps you avoid duplicated code when you find top words for the different subsets of the data. There are different ways to structure it but one way is to feed it subsetted data (so data subsetted to one topic etc.) and for it to get the top words for that subset.


```python
def create_dtm(list_of_strings, metadata):
    vectorizer = CountVectorizer(lowercase=True)
    dtm_sparse = vectorizer.fit_transform(list_of_strings)
    dtm_dense_named = pd.DataFrame(dtm_sparse.todense(), 
                                   columns=vectorizer.get_feature_names_out()) 
    metadata_clean = metadata.copy()
    if 'index' in metadata_clean.columns:
        metadata_clean = metadata_clean.drop(columns=['index'])
    dtm_dense_named_withid = pd.concat([metadata_clean.reset_index(drop=True), dtm_dense_named], axis=1)
    dtm_dense_named_withid = dtm_dense_named_withid.loc[:, ~dtm_dense_named_withid.columns.duplicated()]
    return dtm_dense_named_withid
```

```python
# 2.2 Create a document-term matrix and get top words

# A.
metadata = doj_subset_wscore[["id", "compound", "topics_clean"]]
dtm = create_dtm(doj_subset_wscore["processed_text"], metadata)

# B.
def get_topwords(subset_df, top_n=10):
    word_data = subset_df.drop(columns=["id", "compound", "topics_clean"], errors="ignore")
    return word_data.sum().sort_values(ascending=False).head(top_n)

# C.
top_5 = dtm[dtm["compound"] >= dtm["compound"].quantile(0.95)]
print("Top 10 words in top 5% most positive sentiment:")
print(get_topwords(top_5))

# D.
bottom_5 = dtm[dtm["compound"] <= dtm["compound"].quantile(0.05)]
print("\nTop 10 words in bottom 5% most negative sentiment:")
print(get_topwords(bottom_5))

# E.
for topic in dtm["topics_clean"].unique():
    topic_subset = dtm[dtm["topics_clean"] == topic]
    print(f"\nTop 10 words for topic: {topic}")
    print(get_topwords(topic_subset))
```

**Hints and Resources**:

```python
# your code here
from gensim import corpora, models

target_labels = ["civil rights", "hate crimes", "project safe childhood"]
regex_filter = "|".join(target_labels)
doj_filtered = doj_subset_wscore[doj_subset_wscore["topics_clean"].str.contains(regex_filter, case=False, regex=True)].copy()

print(f"{len(doj_filtered)} documents matched the topic filter.")

token_lists = doj_filtered["processed_text"].apply(lambda x: x.split())
token_lists = [tokens for tokens in token_lists if len(tokens) > 0]

if not token_lists:
    print("No documents with usable tokens.")
else:
    dictionary = corpora.Dictionary(token_lists)
    corpus = [dictionary.doc2bow(tokens) for tokens in token_lists]

    lda_model = models.LdaModel(
        corpus=corpus,
        id2word=dictionary,
        num_topics=3,
        random_state=42,
        passes=10
    )

    topic_names = {
        0: "project safe childhood",
        1: "hate crimes",
        2: "civil rights"
    }

    for idx, topic in lda_model.show_topics(num_topics=3, num_words=15, formatted=False):
        words_only = [word for word, _ in topic]
        print(f"\n{topic_names[idx].title()}:\n" + ", ".join(words_only))
```

```python
## your code here 

# PART A
def create_bigram_onedoc(processed_text):
    words = processed_text.split()
    bigrams = [f"{w1}_{w2}" for w1, w2 in zip(words, words[1:])]
    return " ".join(bigrams)

doj_subset_wscore["processed_text_bigrams"] = doj_subset_wscore["processed_text"].apply(create_bigram_onedoc)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

texts = doj_subset["contents"].tolist()
ids = doj_subset["id"].tolist()

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(texts)

cosine_sim = cosine_similarity(tfidf_matrix)

np.fill_diagonal(cosine_sim, 0)
top_pairs = np.dstack(np.unravel_index(np.argsort(cosine_sim.ravel())[-4:], cosine_sim.shape))[0]

for i in range(0, len(top_pairs), 2):
    idx1, idx2 = top_pairs[i]
    print(f"\nPair {i//2 + 1}:")
    print(f"ID {ids[idx1]}:")
    print(texts[idx1][:500], "...")
    print(f"\nID {ids[idx2]}:")
    print(texts[idx2][:500], "...")
    print(f"\nSimilarity score: {cosine_sim[idx1][idx2]:.4f}")
```
